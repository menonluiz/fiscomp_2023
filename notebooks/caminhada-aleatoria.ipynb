```python
import numpy as np 
import matplotlib.pyplot as plt
import random
```

```python
TamanhoDePasso = .5
NumeroDePassos = 5000
x = 0
ListaX = np.zeros(NumeroDePassos)
for i in range(NumeroDePassos):
    r = random.random()
    if r > .5:
        x = x + TamanhoDePasso
    else:
        x= x - TamanhoDePasso
    ListaX[i] = x
EixoX = np.arange(0,TamanhoDePasso*NumeroDePassos,TamanhoDePasso)
plt.xlabel("N*dx")
plt.ylabel("posição x")
plt.plot(EixoX,ListaX)
plt.show()
```

```python
NumeroDeTrajetorias = 10000
TabelaX = np.zeros((NumeroDeTrajetorias,NumeroDePassos))
for j in range(NumeroDeTrajetorias):
    x=0
    for i in range(NumeroDePassos):
        
        r = random.random()
        if r > .5:
            x = x + TamanhoDePasso
        else:
            x= x - TamanhoDePasso
        TabelaX[j,i] = x

VisualizarTrajetorias=4
Media = np.zeros(NumeroDePassos)

for i in range(NumeroDePassos): 
    Media[i] = np.mean(TabelaX[:,i])

EixoX = np.arange(0,TamanhoDePasso*NumeroDePassos,TamanhoDePasso)
for j in range(VisualizarTrajetorias):
    plt.plot(EixoX,TabelaX[j,:] )
plt.plot(EixoX,Media,c="black")
plt.xlabel("N*dx")
plt.ylabel("posição x")
plt.show()
```

```python
def gauss(x,N,TamanhoDePasso):
    sigma2 = N*TamanhoDePasso**2
    return np.exp(-x**2/(2*sigma2))/np.sqrt(2*np.pi*sigma2)
```

```python
TempoDiscreto = 4000 #tempo em que queremos visualizar o histograma
xx = np.linspace(-300,300,1000)
plt.hist(TabelaX[:,TempoDiscreto], bins=100,density=True,range=(-300,300))
plt.plot(xx,gauss(xx,TempoDiscreto ,TamanhoDePasso))
plt.xlabel("posição x")
plt.ylabel("densidade de probabilidade")
plt.show()
```
