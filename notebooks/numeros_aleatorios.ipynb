```python
num = int(1e4)
r1=np.zeros(num)
for i in range(num):
    r1[i]=2*random.random()-1
    

r2=np.random.uniform(-1,1,num) # ja cria um vetor com "num" elementos

r3=np.zeros(num)

def gerador(x,a,c,m):
    return (x*a +c)%m
    
I =158;a=7**5;m=2**31 -1;c=0
for i in range(1,num):
    I=gerador(I,a,c,m)
    r3[i]=I/m
```

```python
#calculo de autocorrelo
deslocamento_max = 100
auto_correlacao1 = np.zeros(deslocamento_max)
auto_correlacao2 = np.zeros(deslocamento_max)
auto_correlacao3 = np.zeros(deslocamento_max)

media_r1 = 0
media_r2 = 0
media_r3 = 0

media2_r1 = 0
media2_r2 = 0
media2_r3 = 0

for i in range(len(r1)): 
    media_r1= media_r1 + r1[i]/len(r1)
    media_r2= media_r2 + r2[i]/len(r2)
    media_r3= media_r3 + r3[i]/len(r3)

    media2_r1= media2_r1 + (r1[i]**2)/len(r1)
    media2_r2= media2_r2 + (r2[i]**2)/len(r2)
    media2_r3= media2_r3 + (r3[i]**2)/len(r3)
    

sigma2_r1 = media2_r1 - media_r1**2
sigma2_r2 = media2_r2 - media_r2**2
sigma2_r3 = media2_r3 - media_r3**2
    
for i in range(deslocamento_max):
    for j in range(i, len(r1)):
        auto_correlacao1[i] = auto_correlacao1[i] + (r1[j] - media_r1) * (r1[j - i] - media_r1)
        auto_correlacao2[i] = auto_correlacao2[i] + (r2[j] - media_r2) * (r2[j - i] - media_r2)
        auto_correlacao3[i] = auto_correlacao3[i] + (r3[j] - media_r3) * (r3[j - i] - media_r3)

auto_correlacao1 = auto_correlacao1 / (sigma2_r1 * len(r1))
auto_correlacao2 = auto_correlacao2 / (sigma2_r2 * len(r2))
auto_correlacao3 = auto_correlacao3 / (sigma2_r3 * len(r3))
```

```python
plt.plot(auto_correlacao1)
plt.plot(auto_correlacao2)
plt.plot(auto_correlacao3)
plt.show()
```
