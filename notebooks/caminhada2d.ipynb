```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
numero_de_passos = 10000
delta_l = .5
x=np.zeros(numero_de_passos)
y=x.copy()
for i in range (1,numero_de_passos): 
    x[i]=x[i-1]
    y[i]=y[i-1]
    r = np.random.random()
    if r < .25:
        x[i]=x[i-1]+delta_l
    elif (r>=.25)*(r<.5):
        x[i]=x[i-1]-delta_l
    elif(r>=.5)*(r<.75):
        y[i]=y[i-1]+delta_l
    else:
        y[i]=y[i-1]-delta_l
```

```python
plt.plot(y,x)
plt.title("caminhada Aleatria Bidimensional")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
